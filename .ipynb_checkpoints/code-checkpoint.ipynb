- **About**:

Nowadays security is one of the most important factor in digital world. To send data encoded is a method which helps us have security, even a little! There are different methods to encode text data and one of the regular ways is replacement method. Its operation is based on an unknown key which is mapped on alphabets; each character mapped to another one in key. Text data are being encoded by this way and reciever side only can decode it. In this assignment, I try to decode an encoded text with unknown key. It is obvious that to consider all possible states and permutations of 26 characters of English alphabet is impossible since we will have 26! states. So to use genetic algorithm will be very helpful.



- **Problem Modeling**:

To use genetic algorithm, we should model the problem and define chromosomes, mating process, and other fundaments.

> **1.Genes**:
> Genes in this model can have multiple values -Not just 0 and 1-. Every gene can have a value of alphabets: a to z.
>
>**2.Chromosomes**:
> In this problem, a permutation of 26 characters of English alphabets -which is indeed the key.- will be considered as a chromosome.
>
>**3.Fitness Function**:
> According to dictionary we have, the purpose of problem is to find a key by which all words are mapped to meaningful ones. So words which are mapped to meaningful words by a special key, increase fitness of that chromosome. However to consider total number of characters of these words will be way closer to reality. So fitness value of a chromosome will be total number of characters of words which are mapped to meaningful ones.
>
>**4.Genetic Operators**:
>In this section, operation of mutation and cross-over functions are described. More analysis about their functionality will be explained later.
>
>>- **Mutation**:
>>Since every value of genes is unique in this problem, at least two genes will interfere in every mutation operation. On each function call, MUATATION$\times$2 random number in range (0,25) will be selected and their locations will changed two-by-two. MUTATION amount is arbitrary and dependent on problem. To explain more, consider following example. Suppose we have a key like below and MUTATION = 1. Two selected random numbers are 5 and 12:
>>
>>| | | | | | | | | | | | | | | | | | | | | | | | | | | | 
>>|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|-|
>>|**Before mutation**|q|w|e|r|t|y|u|i|o|p|a|s|d|f|g|h|j|k|l|z|x|c|v|b|n|m|
>>|**After mutation** |q|w|e|r|t|d|u|i|o|p|a|s|y|f|g|h|j|k|l|z|x|c|v|b|n|m|
>>
>>
>>- **Cross-over**:
>>Since every value of genes is unique, we can not simple methods of cross-over like single point or multipoint and should consider uniqueness of values. To do this, during every cross-over operation and generating two children, following steps are done:
>>>**1.** A random number in range (0,15) will be selected, say n.
>>>
>>>**2.** A substring of first chromosome is selected which is from nth character to n+10th one. The substring will be placed in first children exactly in the same position it has in first chromosome.
>>>
>>>**3.** After selecting substring, from first index to 25th, I check whether the index has been filled or not. If it has not, the character with the same index in the second string will be selected for this index. If the character had not been placed in string before, it will be placed, otherwise the index will be left empty.
>>>
>>>**4.** After iterating all characters of second chromosome, every empty characters will be filled with remained and unselected characters.
>>>
>>>**5.** Above steps will be done for second child too.
>>>
>>To explain more, consider following example. Suppose we have two chromosomes: Chro1 and Chro2. And selected random number is 8:

>>
>>
>**5.Final State**:
>Clearly final step will be the key by which all words are mapped to meaningful words.


- **Considerations and Explanations**:

Below came some explanations about consideration which is taken during implementation:

>- **Defining Dictionary**: As wanted in problem, we need a reference to check fitness of each chromosome. This can be satisfied by creating a dictionary from a text which contains all words which can come in a file. This file is given. So we just need to eliminate all white spaces and ponctuation marks and save every single word in a set. By this way our dictionary will be ready to use. Data cleaning in this problem is just limited to eliminating unnecessary characters.
>
>
>- **Population**: Based on problem, size of population can change. The higher size, the slower mating process. As well as this, higher size of a special limit will be useless and may lead us into a local minimum since it searches and contains larger search space. In this problem 250 can be logical population size. with population more than this amount (say 1000) we need more time.
>>
>>**Idea of variable population size**: In my opinion and after testing different size of population and static and variable population size, to have growing population can not be a good idea. Since numbers of mating process can be very much and if we increase population's size after every offspring generation, it gets very large which can cause us get stuck in local minimum. As well as this the closer we get to final state, the more limited state space we should have. As a result it may not be very useful.
>>
>
>- **Parent Selection**: To select parents for next generation is one of important and affective steps. To assign a probability to each chromosome base on its fitness function and select parent chromosoms based on it can be helpful. By this way better chromosoms, which have higher fitness amount, will be selected more than weaker ones. Choice function in numpy library does this. Defined probability for each chromosome is: 
>>
>>$\mathrm{P}=\frac{\mathrm{fitness\;of\;current\;chromosome}}{\Sigma\;\mathrm{chromosomes'\;fitness}}$ 
>
>- **Creating Next Generation**: There are different ways to create next generation. In this problem, 10% of best chromosomes are always going to next generation directly and other 90% are results of mating. During mating operation, parents are selected as discussed above. After cross-over operation, mutation will be applied on each children with probability 0.01. It is necessary to avoid getting stuck in local minimum. This procedure will continue until the size of new population reaches 250.
>
>
>- **Mutation versus Cross-over**: There should be a resonable trade off between usage of cross-over and mutation operations and we can not use them lonel. Mutation helps algorithm maintain the diversity in population to avoid the premature convergence. But if we use mutation with no cross-over, algorithm will always have a random population and can never improve toward the best solution. Because cross-over helps algorithm to create new chromosomes from which are better. By this way algorithm can get closer to chromosomes with better fitness and search space converges into final answer. But it is noticable that if we use cross-over without mutation, algorithm can not create any new chromosomes and it just works on initial population which may cause getting stuck in local minimum with no correct answer.
>As a result, resonable trade off between usage of cross-over and mutation operations would be the best choice in genetic algorithms. So degree of these can affect accuracy and selecting affective cross-over operation can increase accuracy more.
>
>
>- **How to deal with local minimums?**: Even though we consider all aspects of problem, we may face premature convergence. It can have different reasons and one of them is the behavior of the function. The function algorithm is applied on, may have many local minimums. This causes algorithm getting stuck in them. In this situation, to change probability of mutation and increase number of random elements can be helpful. In some cases we should be careful that the size of input data may cause premature convergence.
>


- **Failed Solutions**:

Below came different solutions for these problem which were not as good as current one.

>**1. Different cross-over implementation**: The idea was based on random selection for each character of the key. for every character, a random number had been generated. If random number was less than 0.5, character had been selected from first chromosome, otherwise it had been selected from second chromosome. Duplicated character had not been selected and left empty. Finally a random selection between characters ,which had not been placed in new child, had been applied on empty cells. This idea was not good enough, because it had randomness very much and prevents population converge into final answer sooner.
>
>**2. Different parent selection**: The idea was again based on randomness. In this method, first 50% of best chromosomes had same chance to be selected for mating. However it is obvious that it is better to have better chromosomes participate in mating operation, because children will have more chance to get closer to correct answer. So this method faced failure.
>
>**3. Different generation creation**: In this method 10% of current generation had been inserted in new population directy, 80% had been created by cross-over operation and other remaining 10% had been created by random selection. Like first method, its generation was moslty based on randomness, so could not converge into final answer as soon as now. In current solution randomness is supported by mutation with probability 0.01.

- **Implementation**:

```python
%%time
import re
import random
from numpy.random import choice

POPULATION_SIZE = 250
MUTATION = 1
MUTATION_PROB = 0.01
genes = "abcdefghijklmnopqrstuvwxyz"

class Decoder:
    dictionary = set()
    population = {}
    encodedText = ""
    bestRank = 0
    encodedList = []

    def __init__(self, encodedText):
        self.encodedText = encodedText
        self.encodedList = self.cleanData(encodedText)
        self.bestRank = sum(len(w) for w in self.encodedList)
        self.dictionary = self.createDictionary();
        self.initiateRandomPop()
        
    def createDictionary(self):
        rawData = open("./Attachment/global_text.txt").read()
        return(set(self.cleanData(rawData)))
    
    def cleanData(self, rawData):
        rawList = re.split('[^a-zA-Z]',rawData)
        rawList = list(filter(None, rawList))
        return rawList
    
    def initiateRandomPop(self):
        while(len(self.population) != POPULATION_SIZE):
            newChromosome = ''.join(random.sample(genes, len(genes)))
            rank = self.calFitness(newChromosome)
            self.population[newChromosome] = rank
        
    def calFitness(self,chromosome):
        genesUp = genes.upper()
        chromosomeUp = chromosome.upper()
        rank = 0
        keyMapper= {}
        for i in range(26):
            keyMapper[chromosome[i]] = genes[i]
            keyMapper[chromosomeUp[i]] = genesUp[i]
        for word in self.encodedList:
            temp = ""
            for c in word:
                temp += keyMapper[c]
            if temp in self.dictionary:
                rank += len(word)
        return rank
    
    def mutate(self, chromosome):
        chromosome = list(chromosome)
        indexes = random.sample(range(0,25), MUTATION*2)
        for i in range(MUTATION):
            temp = chromosome[indexes[2*i]]
            chromosome[indexes[2*i]] = chromosome[indexes[2*i+1]]
            chromosome[indexes[2*i+1]] = temp
        return ''.join(chromosome)
    
    def crossover(self, chromosome1, chromosome2):
        firstChromosome = []
        secondChromosome = []
        firstChroIdx = random.randint(0,15)
        firstCut = list(chromosome1)[firstChroIdx:firstChroIdx+10]
        secondCut = list(chromosome2)[firstChroIdx:firstChroIdx+10] 
        i1 = 0
        i2 = 0
        while(i1 != 26 or i2 != 26):
            if(i1 < 26):
                if(i1 == firstChroIdx):
                    firstChromosome += firstCut
                    i1 += 10
                if(chromosome2[i1] not in firstCut):
                    firstChromosome += chromosome2[i1]
                else:
                    firstChromosome += [""]
                i1 += 1
            if(i2 < 26):
                if(i2 == firstChroIdx):
                    secondChromosome += secondCut
                    i2 += 10
                if(chromosome1[i2] not in secondCut):
                    secondChromosome += chromosome1[i2]
                else:
                    secondChromosome += [""]
                i2 += 1
        firstEmpty = [i for i in range(len(firstChromosome)) if firstChromosome[i] == ""]
        secondEmpty = [i for i in range(len(secondChromosome)) if secondChromosome[i] == ""]
        j1 = 0
        j2 = 0
        for i in range(26):
            if(chromosome2[i] not in firstChromosome):
                firstChromosome[firstEmpty[j1]] = chromosome2[i]
                j1+=1
            if(chromosome1[i] not in secondChromosome):
                secondChromosome[secondEmpty[j2]] = chromosome1[i]
                j2+=1
        return (''.join(firstChromosome),''.join(secondChromosome))
    
    def findKey(self):
        sortedPop = sorted(self.population.items(), key = lambda kv:(kv[1], kv[0]))
        while(sortedPop[-1][1] != self.bestRank):
            newPop = dict(sortedPop[int(10*POPULATION_SIZE/100):POPULATION_SIZE])
            scoreSum = sum(n[1] for n in sortedPop)
            if(not(scoreSum)):
                scoreSum = 1

            while(len(newPop) != POPULATION_SIZE):
                chromosomes = choice([p[0] for p in sortedPop], 2, [p[1]/scoreSum for p in sortedPop])
                newChromosomes = self.crossover(chromosomes[0], chromosomes[1])
                firstChromosome = newChromosomes[0]
                secondChromosome = newChromosomes[1]
                if(random.random() < MUTATION_PROB):
                    firstChromosome = self.mutate(firstChromosome)
                if(random.random() < MUTATION_PROB):
                    secondChromosome = self.mutate(secondChromosome)
                newPop[firstChromosome] = self.calFitness(firstChromosome)
                if(len(newPop) != POPULATION_SIZE):
                    newPop[secondChromosome] = self.calFitness(secondChromosome)
            self.population = newPop
            sortedPop = sorted(self.population.items(), key = lambda kv:(kv[1], kv[0]))
        return(sortedPop[-1][0])
    
    def decode(self):
        finalText = ""
        decodeKey = self.findKey()
        genesUp = genes.upper()
        decodeKeyUp = decodeKey.upper()
        keyMapper= {}
        for i in range(26):
            keyMapper[decodeKey[i]] = genes[i]
            keyMapper[decodeKeyUp[i]] = genesUp[i]
        for char in self.encodedText:
            if(char in keyMapper):
                finalText += keyMapper[char]
            else:
                finalText += char
                
        return(finalText)

                
encodedText = open("./Attachment/encoded_text.txt").read()
d = Decoder(encodedText)
decodedText = d.decode()
print(decodedText)
```
